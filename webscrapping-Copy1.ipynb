```python
import bs4
from urllib.request import urlopen as uReq # connecting to webclient
from bs4 import BeautifulSoup as soup
```

```python
myurl="https://ec.europa.eu/esco/portal/skill?uri=http%3A%2F%2Fdata.europa.eu%2Fesco%2Fskill%2F000f1d3d-220f-4789-9c0a-cc742521fb02&conceptLanguage=en&full=true#&uri=http://data.europa.eu/esco/skill/000f1d3d-220f-4789-9c0a-cc742521fb02"  #webpage url to scrap
uClient=uReq(myurl)  # loading the url
```

```python
page_html=uClient.read() #read and store the webpage
uClient.close() #close the html page as it is downloaded from internet and it is big
```

```python
page_soup=soup(page_html,"html.parser")  #html parsing
```

```python
import re
def clean_string(next5):
    x=re.sub('\s+ ',',',next5)
    return x
```

```python
# extracting data under each header
headers=containers[0].findAll('h2')
l=[]

next0=headers[0].find_next_sibling(['pre','h2']).text
next1=headers[1].find_next_sibling(['ul','li','p','h2']).text.strip()
next2=headers[2].find_next_sibling(['ul','li','p','h2']).text.strip()
next3=headers[3].find_next_sibling(['ul','li','a','h2']).text.strip()
next4=headers[4].find_next_sibling(['ul','li','a','h2']).text.strip()
next5=headers[5].find_next_sibling(['ul','li','a','h2']).text.strip()
next6=headers[6].find_next_sibling(['p','h2']).text
next7=headers[7].find_next_sibling(['p','h2']).text
next=[next0,next1,next2,next3,next4,next5,next6,next7]
for i in range (0,8):
    l1=(headers[i].text,clean_string(next[i]))
    l.append(l1)
print(l)
```

```python
#Create dataframe

import pandas as pd
pd.set_option('max_colwidth', -1)
df=pd.DataFrame(l,columns=["Heading","Subheadings"])
df
```
